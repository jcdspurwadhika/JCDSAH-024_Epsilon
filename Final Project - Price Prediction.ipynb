## **Business Problem Understanding**  
### **Context**
Di era digital ini untuk orang yang ingin membeli mobil sudah sangat dimudahkan dengan adanya internet. Kemudahan yang didapat adalah bisa mencari dan membandingkan mobil sesuai impian atau pilihan mereka tanpa datang lebih dulu. Salah satunya platform website penjualan mobil bekas di Arab Saudi yaitu www.syarah.com. 

Penjual pun sangat dimudahkan dalam memasarkan mobil yang dijualnya untuk menjangkau pasar yang lebih luas. Namun, untuk penjual itu menjadi sebuah tantangan dalam menentukan harga yang sesuai. Dimana kalau harga tidak sesuai atau contoh terlalu mahal atau terlalu rendah, berpotensi menyebabkan pendapatan yang tidak maksimal dan waktu penjualan yang lama. 

### **Problem Statement**
Bagaimana membangun sistem estimasi harga mobil bekas yang mampu merepresentasikan nilai pasar berdasarkan karakteristik kendaraan, sehingga membantu penjual menentukan harga yang kompetitif dan meningkatkan peluang terjadinya transaksi?

### **Stakeholder**
Stakeholder dalam analisis ini meliputi:
1. Platform syarah.com, sebagai penyedia marketplace kendaraan mobil bekas yang berkepentingan dalam meningkatkan jumlah transaksi dan menjaga kredibilitas harga di platform.

2. Penjual kendaraan, yang membutuhkan referensi harga agar tidak mengalami lost profit akibat underpricing maupun keterlambatan penjualan akibat harga yang terlalu mahal atau overprice.

3. Calon pembeli, yang diuntungkan dengan harga yang lebih transparan dan sesuai dengan kondisi pasar.

### **Goal**
Tujuan dari analisis ini adalah membangun model machine learning berbasis data listing mobil bekas untuk memprediksi dan mengestimasi harga pasar berdasarkan spesifikasi kendaraan. Model ini dirancang sebagai sistem rekomendasi harga yang dapat diintegrasikan ke dalam platform marketplace, sehingga membantu penjual menentukan harga yang lebih optimal dan kompetitif sesuai kondisi pasar.

Implementasi model ini diharapkan memberikan beberapa manfaat strategis, antara lain:

- Meningkatkan Profit  
Dengan estimasi harga yang lebih akurat, penjual dapat menghindari underpricing yang berpotensi menyebabkan kehilangan keuntungan. Harga yang sesuai dengan nilai pasar memungkinkan penjual memaksimalkan profit tanpa kehilangan daya saing.

- Mempercepat Waktu Penjualan  
Penetapan harga yang terlalu tinggi sering kali membuat mobil sulit terjual. Rekomendasi harga berbasis data membantu menjaga harga tetap kompetitif, sehingga mempercepat proses transaksi.

- Meningkatkan Kepercayaan Pengguna terhadap Platform  
Fitur seperti “Suggested Market Price” meningkatkan transparansi harga dan memberikan referensi objektif kepada penjual maupun pembeli, sehingga memperkuat kredibilitas platform.

- Meningkatkan Conversion Rate Platform  
Harga yang lebih rasional dan sesuai pasar meningkatkan peluang terjadinya transaksi. Hal ini berpotensi meningkatkan volume penjualan dan revenue platform, terutama jika model bisnis berbasis komisi.

### **Analytic Approach**  
Data akan dianalisis menggunakan pendekatan Exploratory Data Analysis (EDA) untuk menemukan pola dan hubungan antar fitur yang memengaruhi harga mobil bekas. Selanjutnya, akan dibangun model Machine Learning berbasis regresi untuk memprediksi harga mobil di platform syarah.com. Pendekatan regresi dipilih karena target variabel yang ingin diprediksi adalah harga yaitu sebuah nilai numerik kontinu, bukan kategori.  
Penggunaan Machine Learning diperlukan karena penetapan harga secara manual rentan terhadap bias dan ketidakakuratan. Dengan Machine Learning, model dapat belajar dari data listing historis untuk menghasilkan estimasi harga yang lebih objektif dan berbasis data, sehingga membantu penjual menentukan harga yang kompetitif di platform syarah.com.

### **Evaluation Metric**
Model regresi dievaluasi menggunakan beberapa metrik berikut untuk mengukur performa prediksi harga:

1. MAE (Mean Absolute Error)  
Mengukur rata-rata selisih absolut antara harga aktual dan harga prediksi dalam satuan SAR.
MAE dipilih karena mudah diinterpretasikan oleh stakeholder bisnis dan secara langsung merepresentasikan rata-rata kesalahan harga yang terjadi pada setiap prediksi.

2. RMSE (Root Mean Squared Error)  
Mengukur akar dari rata-rata kuadrat error.
RMSE memberikan penalti lebih besar terhadap error yang ekstrem (outlier), sehingga membantu mengidentifikasi apakah model menghasilkan kesalahan besar pada mobil dengan harga sangat tinggi atau sangat rendah.

3. MAPE (Mean Absolute Percentage Error)  
Mengukur rata-rata persentase kesalahan prediksi terhadap harga aktual.
MAPE membantu memahami besar kesalahan dalam bentuk persentase, sehingga lebih intuitif untuk membandingkan performa model pada berbagai rentang harga.

4. R² Score  
Mengukur seberapa besar variasi harga mobil yang dapat dijelaskan oleh fitur-fitur dalam model.
Nilai R² yang mendekati 1 menunjukkan bahwa model mampu menjelaskan sebagian besar variasi harga di dalam dataset.

Visualisasi Car Mileage saat ini tidak informatif dikarenakan adanya extreme outlier pada angka 20.000.000 km yang menarik skala sumbu Y terlalu jauh ke atas. Data ini terindikasi kuat sebagai data eror dan harus dibersihkan agar distribusi jarak tempuh mobil yang sebenarnya dapat terlihat dengan jelas.

```python
# Menghitung IQR untuk kolom Mileage dan menghapus outlier berdasarkan IQR
Q1 = df["Mileage"].quantile(0.25)
Q3 = df["Mileage"].quantile(0.75)
IQR = Q3 - Q1

upper_bound = Q3 + 1.5 * IQR

df = df[df["Mileage"] <= upper_bound]
```

Selanjutnya kita akan menghapus outlier tersebut menggunakan metode Interquartile Range (IQR) untuk menentukan rentang jarak tempuh yang dianggap wajar. Alasan utama penggunaan metode IQR adalah karena sifatnya yang robust terhadap outlier. Tidak seperti nilai rata-rata (mean) yang akan bergeser jauh akibat angka 20.000.000 km, IQR hanya berfokus pada 50% data di bagian tengah, sehingga batas pembersihan data menjadi lebih objektif dan tetap mencerminkan mayoritas populasi mobil yang sebenarnya.

```python
plt.figure(figsize=(5, 4))
plt.boxplot(df["Mileage"])
plt.title("Distribution of Car Mileage")
plt.ylabel("Mileage")
plt.show()
```

Sekarang boxplot sudah dapat terlihat jelas distribusinya. Di sini kita menghapus outlier pada feature Mileage dengan menggunakan metode IQR dan sekarang outlier 20.000.000 km sudah hilang. Median di sekitar 100.000 km, Q1 45.000 km dan Q3 165.000 yang artinya 50% mobil berada di rentang itu. Ada juga nilai yang di atas 350.000 km tapi masih realistis untuk mobil tua yang kmnya sudah banyak sehingga tidak dihapus.

## **Data Preparation**

```python
# Import library untuk modeling
from sklearn.model_selection import train_test_split, cross_val_score, RandomizedSearchCV, KFold

import category_encoders as ce
from sklearn.preprocessing import OneHotEncoder, RobustScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from xgboost.sklearn import XGBRegressor
from sklearn.compose import TransformedTargetRegressor

from sklearn.preprocessing import StandardScaler, OrdinalEncoder, OneHotEncoder, FunctionTransformer, TargetEncoder
from sklearn.impute import SimpleImputer

from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error, mean_absolute_percentage_error
```

```python
df = pd.read_csv('UsedCarsSaudiCleaned.csv')
```

```python
df.head()
```

#### **Feature Engineering & Preprocessing Pipeline**

```python
from datetime import datetime
# Tahun mobil
df["Car_Age"] = datetime.now().year - df["Year"]
df.drop(columns=["Year"], inplace=True)
```

Dibuat fitur baru Car_Age yang merepresentasikan usia kendaraan (tahun sekarang dikurangi tahun produksi). Fitur ini digunakan karena harga mobil lebih dipengaruhi oleh umur kendaraan dibandingkan angka tahun produksi secara langsung. Kolom Year kemudian dihapus untuk menghindari redundansi informasi.

```python
bin_col = ["Gear_Type"]
ohe_col = ["Origin", "Color", "Options", "Fuel_Type", "Region"]
high_card_cols = ["Make", "Type"] 
num_col = ["Engine_Size", "Mileage", "Car_Age"]
```

```python
numeric_pipeline = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler",  RobustScaler()),
])

binary_pipeline = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("ordinal", OrdinalEncoder(handle_unknown="use_encoded_value", unknown_value=-1)),
])

ohe_pipeline = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("ohe", OneHotEncoder(handle_unknown="ignore", sparse_output=False)),
])

high_card_pipeline = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("target", TargetEncoder(target_type='continuous', random_state=1)),
])
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ("num",      numeric_pipeline,   num_col),
        ("bin",      binary_pipeline,    bin_col),
        ("ohe",      ohe_pipeline,       ohe_col),
        ("high_card",high_card_pipeline, high_card_cols),
    ],
    remainder='drop'
)
```

#### **Data Splitting**

```python
X = df.drop(['Price'], axis=1)
y = df['Price']
```

```python
X_train, X_test, y_train, y_test= train_test_split(
    X,
    y,  
    test_size=0.2,
    random_state=1)
```

#### **Benchmark Model**

```python
# Define algoritma yang digunakan
lr = LinearRegression()
knn = KNeighborsRegressor()
dt = DecisionTreeRegressor(random_state=1)
rf = RandomForestRegressor(random_state=1)
xgb = XGBRegressor(random_state=1)

# Pemodelan dilakukan dalam skala logaritmik, namun kemudian di-inverse kembali untuk interpretasi
log_lr = TransformedTargetRegressor(lr, func=np.log, inverse_func=np.exp)
log_knn = TransformedTargetRegressor(knn, func=np.log, inverse_func=np.exp)
log_dt = TransformedTargetRegressor(dt, func=np.log, inverse_func=np.exp)
log_rf = TransformedTargetRegressor(rf, func=np.log, inverse_func=np.exp)
log_xgb = TransformedTargetRegressor(xgb, func=np.log, inverse_func=np.exp)


# Kandidat algoritma
model_names = ['Linear Regression', 'KNN Regressor', 'DecisionTree Regressor', 'RandomForest Regressor', 'XGBRegressor']
models = [log_lr, log_knn, log_dt, log_rf, log_xgb]

nilai_mean_rmse, nilai_std_rmse = [], []
nilai_mean_mae,  nilai_std_mae  = [], []
nilai_mean_mape, nilai_std_mape = [], []
nilai_mean_r2, nilai_std_r2 = [], []

# Mencari algoritma terbaik berdasarkan tiap metrics
for name, model in zip(model_names, models):

    crossval = KFold(n_splits=5, shuffle=True, random_state=1)

    estimator = Pipeline([
        ('preprocessing', preprocessor),
        ('model', model)
    ])

    cv_rmse = cross_val_score(estimator, X_train, y_train, cv=crossval,
                                scoring='neg_root_mean_squared_error', error_score='raise')
    cv_mae  = cross_val_score(estimator, X_train, y_train, cv=crossval,
                                scoring='neg_mean_absolute_error', error_score='raise')
    cv_mape = cross_val_score(estimator, X_train, y_train, cv=crossval,
                                scoring='neg_mean_absolute_percentage_error', error_score='raise')
    cv_r2 = cross_val_score(estimator, X_train, y_train, cv=crossval,
                                scoring='r2', error_score='raise')

    nilai_mean_rmse.append(abs(cv_rmse.mean()))
    nilai_std_rmse.append(cv_rmse.std())
    nilai_mean_mae.append(abs(cv_mae.mean()))
    nilai_std_mae.append(cv_mae.std())
    nilai_mean_mape.append(abs(cv_mape.mean()))
    nilai_std_mape.append(cv_mape.std())
    nilai_mean_r2.append(cv_r2.mean())    
    nilai_std_r2.append(cv_r2.std())

    print(f"  {name}: RMSE={abs(cv_rmse.mean()):,.2f} (±{cv_rmse.std():,.2f}) | "
      f"MAE={abs(cv_mae.mean()):,.2f} (±{cv_mae.std():,.2f}) | "
      f"MAPE={abs(cv_mape.mean()):.4f} (±{cv_mape.std():.4f}) | "
      f"R²={cv_r2.mean():.4f} (±{cv_r2.std():.4f})")

cv_results = pd.DataFrame({
    'Model'    : model_names,
    'Mean_RMSE': nilai_mean_rmse,
    'Std_RMSE' : nilai_std_rmse,
    'Mean_MAE' : nilai_mean_mae,
    'Std_MAE'  : nilai_std_mae,
    'Mean_MAPE': nilai_mean_mape,
    'Std_MAPE' : nilai_std_mape,
    'Mean_R2': nilai_mean_r2,
    'Std_R2' : nilai_std_r2,
})
```

```python
cv_results
```

Berdasarkan hasil uji coba terhadap lima algoritma regresi, XGBRegressor dan RandomForest Regressor terpilih sebagai dua model dengan performa terbaik. XGBRegressor menjadi yang utama karena mencatatkan nilai rata-rata R2 tertinggi sebesar 0,805, serta tingkat kesalahan rata-rata (MAE) terendah di angka 13.969. Sementara itu, RandomForest dipilih sebagai model pembanding dengan tingkat stabilitas prediksi yang tinggi (ditandai dengan nilai standar deviasi yang rendah).

#### **Predict Test Set dengan 2 Benchmark Model**

```python
# Benchmark 2 model terbaik
models = {
    'XGB': TransformedTargetRegressor(XGBRegressor(random_state=1), func=np.log, inverse_func=np.exp),
    'RandomForest': TransformedTargetRegressor(RandomForestRegressor(random_state=1), func=np.log, inverse_func=np.exp)
}

score_rmse, score_mae, score_mape, score_r2 = [], [], [], []

# Prediksi pada test set
for name, model in models.items():
    pipeline = Pipeline([
        ('preprocessing', preprocessor),
        ('model', model)
    ])
    pipeline.fit(X_train, y_train)
    y_pred = pipeline.predict(X_test)
    
    if name == 'XGB':
        y_pred_xgb_before_tuning = y_pred

    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    mae  = mean_absolute_error(y_test, y_pred)
    mape = mean_absolute_percentage_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    
    score_rmse.append(rmse)
    score_mae.append(mae)
    score_mape.append(mape)
    score_r2.append(r2)

    print(f"  {name}: RMSE={rmse:,.2f} | MAE={mae:,.2f} | MAPE={mape:.4f} | R²={r2:.4f}")

score_before_tuning = pd.DataFrame(
    {'RMSE': score_rmse, 'MAE': score_mae, 'MAPE': score_mape, 'R2': score_r2},
    index=models.keys()
)
```

```python
score_before_tuning
```

- RMSE & MAE lebih rendah yang artinya Error prediksi XGB lebih kecil.
- MAPE 18.02 artinya rata-rata kesalahan prediksi sekitar 18.02% dari harga asli.

#### **Hyperparameter Tuning**

Karena model XGBoost memiliki performa yang sedikit lebih baik dari RandomForest, maka selanjutnya kita akan menggunakan model XGBoost ini sebagai model akhir. Pada tahap ini, kita akan melakukan hyperparameter tuning pada model XGBoost dengan harapan dapat meningkatkan performa model.

```python
hyperparam_space_xgb = {
    'model__regressor__max_depth'        : [4, 5, 6, 7],
    'model__regressor__learning_rate'    : [0.01, 0.05, 0.1],
    'model__regressor__n_estimators'     : list(np.arange(200, 501)),
    'model__regressor__subsample'        : [0.7, 0.8, 0.9],
    'model__regressor__gamma'            : [0, 0.1, 0.5, 1],
    'model__regressor__colsample_bytree' : [0.6, 0.7, 0.8],
    'model__regressor__reg_alpha'        : [0, 0.1, 0.5, 1.0],
    'model__regressor__reg_lambda'       : [1.0, 2.0, 5.0],
    'model__regressor__min_child_weight' : [3, 5, 10],
}
```

```python
# Benchmark model dengan hyperparameter tuning
xgb = XGBRegressor(random_state=1, verbosity=0)

log_xgb_tuning = TransformedTargetRegressor(
    xgb, func=np.log, inverse_func=np.exp
)

# Membuat algorithm chains
estimator_xgb = Pipeline([
        ('preprocessing', preprocessor),
        ('model', log_xgb_tuning)
        ])

crossval = KFold(n_splits=5, shuffle=True, random_state=1)

# Hyperparameter tuning
random_xgb = RandomizedSearchCV(
    estimator_xgb,
    param_distributions=hyperparam_space_xgb,
    n_iter=100,
    cv=crossval,
    scoring=[
        'neg_root_mean_squared_error',
        'neg_mean_absolute_error',
        'neg_mean_absolute_percentage_error',
        'r2',
    ],
    n_jobs=-1,
    refit='neg_root_mean_squared_error',    
    random_state=1
)
```

```python
# Fitting data training untuk mencari parameter terbaik
random_xgb.fit(X_train, y_train)
```

```python
# Melihat hasil tuning dalam bentuk dataframe. Diurutkan berdasarkan RMSE, MAE & MAPE
pd.DataFrame(random_xgb.cv_results_).sort_values(by=['rank_test_neg_root_mean_squared_error', 'rank_test_neg_mean_absolute_error', 'rank_test_neg_mean_absolute_percentage_error']).head()
```

```python
# Cek skor dan params terbaik
print('XGBoost')
print('Best_score:', random_xgb.best_score_)
print('Best_params:', random_xgb.best_params_)
```

2. **Type**: Menempati posisi kedua. Setelah diproses dengan TargetEncoder, Type merepresentasikan rata-rata harga per model mobil sehingga model langsung mendapat informasi nilai pasar spesifik tiap tipe kendaraan.
3. **Engine_Size**: Kapasitas mesin berada di urutan ketiga. Ukuran mesin seperti 2.5L atau 3.5L menjadi faktor penentu harga yang signifikan karena berkaitan langsung dengan performa dan segmen kendaraan.
4. **Options_Full**: Kelengkapan fitur menjadi pembeda utama antara mobil murah dan mahal. Mobil dengan opsi Full secara konsisten dihargai lebih tinggi di pasar.
5. **Mileage**: Meskipun penting, jarak tempuh berada di bawah Engine_Size. Ini menarik karena menunjukkan bahwa di pasar Arab Saudi, pembeli lebih peduli pada spesifikasi mesin dan kelengkapan fitur daripada angka odometer.
6. **Options_Standard**: Mempertegas bahwa perbedaan antara tipe Standard dan Full adalah informasi kunci bagi model untuk memprediksi harga secara akurat.
7. **Make**: Merk mobil tetap berpengaruh terhadap harga meskipun berada di bawah Type. Ini wajar karena Type lebih spesifik dibanding Make dalam merepresentasikan nilai pasar kendaraan.

## **Conclusion**  
Model yang digunakan untuk menerapkan machine learning dalam memprediksi harga mobil bekas adalah Extreme Gradient Boost (XGB). Cara XGB bekerja adalah model XGB membuat tree layaknya seperti model Random Forest, akan tetapi forest yang dibuat oleh XGB ini dibuat secara sequential dimana tree baru akan dibuat untuk memperbaiki error pada tree sebelumnya. Dalam dataset used cars ini direpresentasikan dengan cara berikut:

- Tree pertama akan membuat prediksi harga secara kasar untuk setiap mobil di dataset.
- Tree berikutnya akan membuat prediksi berdasarkan hasil error dari tree pertama dan seterusnya sebanyak n_estimators pada hyperparameter tuning space.

Berdasarkan analisis dataset 5.235 mobil bekas Arab Saudi, Car_Age terbukti menjadi faktor paling dominan dalam menentukan harga yang artinya semakin tua mobil, semakin rendah nilainya. Type menempati posisi kedua karena merepresentasikan nilai pasar spesifik tiap model kendaraan. Engine_Size memiliki korelasi positif lemah (r = 0.32) namun tetap masuk top 3 feature importance. Hasil ini mengkonfirmasi bahwa usia kendaraan dan tipe model adalah faktor utama yang menentukan harga mobil bekas di pasar Arab Saudi.
